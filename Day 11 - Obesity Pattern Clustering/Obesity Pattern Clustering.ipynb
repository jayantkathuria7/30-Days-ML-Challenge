## Data Preprocessing

```python
df = df.rename(columns={'FAVC': 'frequent_high_calorie_food', 
                   'FCVC': 'eat_vegetables',
                   'NCP': 'num_meals_daily', 
                   'CAEC': 'food_bw_meals', 
                   'SMOKE': 'smoke', 
                   'CH2O': 'water_amt_daily', 
                   'SCC': 'monitor_calories', 
                   'FAF': 'physical_activity_frequency', 
                   'TUE': 'time_spent_on_technology', 
                   'CALC':'alcohol_frequency', 
                   'MTRANS': 'most_used_transportation',
                   'NObeyesdad': 'obesity_level'})
```

```python
df=df.drop_duplicates()
```

```python
df1 = df.drop('obesity_level', axis=1)
```

#### Encoding

```python
categorical_cols = []
numerical_cols=[]
for col in df1.columns:
    if df[col].dtype==object:
        categorical_cols.append(col)
        print(df[col].value_counts())
        print()
    else:
        numerical_cols.append(col)
```

```python
categorical_cols
```

```python
numerical_cols
```

```python
gender_mapped = {'Male':1, 'Female':0}
family_history_mapped = {'yes':1, 'no':0}
high_calorie_food = {'yes':1, 'no':0}
food_bw_meals_mapped = {'Sometimes':1, 'Frequently':2, 'Always':3, 'no':0}
smoke_mapped = {'yes':1, 'no':0}
monitor_calories_mapped = {'yes':0,'no':1}
alcohol_frequency_mapped = {'Sometimes':1, 'Frequently':2, 'Always':3, 'no':0}
transportation_mapped = {'Public_Transportation':2, 'Automobile':4, 'Walking':0, 'Motorbike':3, 'Bike':1}
```

```python
mappings = [gender_mapped, family_history_mapped, high_calorie_food, food_bw_meals_mapped, smoke_mapped, 
            monitor_calories_mapped, alcohol_frequency_mapped, transportation_mapped ]
```

```python
for i, col in enumerate(categorical_cols):
    df1[col] = df1[col].map(mappings[i])
```

```python
df1.head()
```

#### Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df1)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
pca.fit(scaled_data)
```

```python
W = pca.components_.T
W
```

```python
pd.DataFrame(W, index=df1.columns, columns=['col1','col2','col3'])
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
sns.barplot(x = list(range(1,4)), y = pca.explained_variance_, palette = 'GnBu_r')
plt.xlabel('i')
plt.ylabel('Lambda i')
```

```python
PCA_ds = pd.DataFrame(pca.transform(df1), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
#A 3D Projection Of Data In The Reduced Dimension
x =PCA_ds["col1"]
y =PCA_ds["col2"]
z =PCA_ds["col3"]
#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

```python
# Elbow method to find numbers of clusters to make.
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans, AgglomerativeClustering
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_ds)
Elbow_M.show()
```

```python
#Initiating the Agglomerative Clustering model
AC = AgglomerativeClustering(n_clusters=4)
# fit model and predict clusters
yhat_AC = AC.fit_predict(PCA_ds)
PCA_ds["Clusters"] = yhat_AC
#Adding the Clusters feature to the orignal dataframe.
df1["Clusters"]= yhat_AC
```

```python
#Plotting the clusters
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=PCA_ds["Clusters"], marker='o', cmap = 'viridis' )
ax.set_title("The Plot Of The Clusters")
plt.show()
```

```python
#Plotting countplot of clusters
pl = sns.countplot(x=df1["Clusters"], palette= 'viridis')
pl.set_title("Distribution Of The Clusters")
plt.show()
```
