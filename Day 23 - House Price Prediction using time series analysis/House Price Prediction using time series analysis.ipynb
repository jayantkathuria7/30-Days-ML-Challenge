```python
df = df.backfill(axis=1)
df
```

```python
df.columns
```

```python
df.fillna(method='ffill', inplace=True) 
```

```python
# Convert dataset to time-series format
regions = df[['RegionName']]
time_series_data = df.iloc[:, 5:].T
```

```python
time_series_data.index = pd.to_datetime(time_series_data.index)
time_series_data.columns = df['RegionName']
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Forecasting for all regions
forecast_steps = 12
forecast_results = {}

for region in time_series_data.columns:
    try:
        series = time_series_data[region].dropna()
        model = ARIMA(series, order=(5, 1, 0))  # Adjust order as needed
        model_fit = model.fit()
        forecast = model_fit.forecast(steps=forecast_steps)
        forecast_index = pd.date_range(start=series.index[-1], periods=forecast_steps+1, freq='M')[1:]
        forecast_results[region] = pd.Series(forecast, index=forecast_index)
    except:
        print(f"Skipping {region} due to insufficient data")

# Plot results for a few sample regions
plt.figure(figsize=(12, 6))
for region in list(time_series_data.columns[:5]):  # Plot only first 5 regions
    if region in forecast_results:
        plt.plot(time_series_data[region], label=f'Actual: {region}')
        plt.plot(forecast_results[region], linestyle='dashed', label=f'Forecast: {region}')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Time Series Forecasting for Multiple Regions using ARIMA')
plt.legend()
plt.show()
```
